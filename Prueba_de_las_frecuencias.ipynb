**Prueba de las frecuencias**

Se utilizara el metodo de las frecuancias para poder ver si los numeros son aleatorios, donde obtendremos nuestro valor etadisticos

$$VS=\frac{(F_o-F_e)^2}{F_e}$$

$F_e=$ Frecuencia esperada

$F_o=$ Frecuencia observada

donde la suma de este valor debe ser menor al valor critico de nuestros numeros.

$$∑VS<V_C$$
