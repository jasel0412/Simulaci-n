# Prueba de las rachas
A continuación se realizara y ejecutara la pruba de las rachas, la cual consiste en comparar el número anterior con el que estmos en posición, denotaremos nuestro primer número posición como $n_1$,  nuestro número anteriror como $n_0$, entonces en la lista de números que tendremos hasta $k$ números. Nuestra prueba consiste en que si nuestro número posición es mayor al anterior entonces esto nos devuelve un 1, y si nuestro número posición es menor al anterior es 0, generalizando lo anterior.

$$n_k<n_{k-1} \implies 0$$
$$n_k>n_{k-1} \implies 1$$

```python
import math
import numpy as np

s = np.array([
    [0.78961, 0.05230, 0.10699, 0.55877, 0.1415],
    [0.76086, 0.12079, 0.27738, 0.65726, 0.79269],
    [0.80548, 0.82654, 0.29453, 0.20852, 0.42989],
    [0.58518, 0.98611, 0.34488, 0.34358, 0.11537],
    [0.89898, 0.57880, 0.67621, 0.05010, 0.00121],
    [0.28269, 0.73059, 0.70119, 0.18284, 0.49962],
    [0.38618, 0.76910, 0.68334, 0.55170, 0.10850],
    [0.79982, 0.45679, 0.21631, 0.87616, 0.55743],
    [0.58962, 0.33216, 0.03185, 0.61168, 0.09264],
    [0.69623, 0.17028, 0.05475, 0.91512, 0.76262],
    [0.29931, 0.30861, 0.83358, 0.51781, 0.03272],
    [0.57410, 0.26593, 0.85903, 0.43308, 0.35286],
    [0.24000, 0.65559, 0.38507, 0.90829, 0.94187],
    [0.93655, 0.88809, 0.81772, 0.36982, 0.19904],
    [0.54325, 0.62400, 0.09133, 0.41678, 0.33954],
    [0.58244, 0.85853, 0.88752, 0.33729, 0.15506],
    [0.23949, 0.53559, 0.33381, 0.49383, 0.75103],
    [0.19962, 0.65002, 0.74579, 0.79113, 0.63453],
    [0.19147, 0.40644, 0.08128, 0.73435, 0.22724],
    [0.22287, 0.07281, 0.64183, 0.44267, 0.72102]
])

t=[]#creamos una lista vacía para poder guardar los valores

for j in range(s.shape[1]):#movera la columna de los valores
    for i in range(s.shape[0]):#movera la fila de los valores
        n_i = (i + 1) % s.shape[0]
        if s[i, j] > s[n_i, j]:#la condición de la prueba de rachas
            t.append(0)#agrega los valores a la lista
        else:
            t.append(1)#agrega los valores a la lista

cont = 1 #ponemos un contador
print(len(t)) #contamos los valores obtenidos en nuestra prueba de rachas

for z in range(len(t)):#implementamos un ciclo para contar las rachas
    if t[z] != t[(z+1)% len(t)]:#cuando el número posicion de la lista es diferente al siguiente, suma 1 al contador
      cont += 1

print(cont)#imprimimos las rachas obtenidas
```
